```
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('classic')  # Библитотека matplotlib имеет 10 цветовых стилей
sns.set()                 # Установим отрисовку графиков с помощью библиотеки seaborn
import numpy as np
x=np.linspace(start=0, stop=10, num=100) # Сгенерируем 100 чисел в диапазоне от 0 до 10
fig=plt.figure()  # Создадим полотно, на котором будем визуализировать данные. Оно позволит отрисовать несколько графиков в рамках одного пространства
plt.plot(x, np.sin(x), label='Синус')    # Создадим график №1 (sin(x))
plt.plot(x, np.cos(x), label='Косинус')  # Создадим график №2 (cos(x))
leg=plt.legend()  # Выведем легенду на график
plt.title('Кривые синуса и косинуса')
plt.xlabel('Значение х')
plt.ylabel('Значение функции')
```

**Линейный график** состоит из точек, соединённых линией, которая показывает связь между переменными x и y. Ось x обычно содержит временные интервалы, ось y — числовую переменную, изменения которой во времени нужно отследить.

```
sorted_df=df.sort_values(by=['Date', 'Time']) # Сортируем значения сначала по Date, затем по Time
filtered_df=sorted_df[(sorted_df.Branch=='A')&(sorted_df.Payment=='Cash')] # В отсортированном Датасете отфильтруем по филиалу и способу оплаты
plt.figure(figsize=(20,8))  # Зададим размер графика
plt.plot(filtered_df['Invoice ID'], filtered_df['gross income'])
plt.title('Значение валового дохода по филиалу А при денежном способе оплаты')
plt.xlabel('Идентификационный номер (отсортирован по времени)')
plt.ylabel('Валовый доход')
```

**ГИСТОГРАММА**

```
# ИСПОЛЬЗОВАНИЕ MATPLOTLIB.PYPLOT
plt.figure(figsize=(10,6))  # Зададим размер графика
plt.hist(df['Payment'])
plt.title('Количество супермаркетов по способам оплаты')
plt.xlabel('Способ оплаты')
plt.ylabel('Количество супермаркетов')
```

**ИСПОЛЬЗОВАНИЕ SEABORN**

**Гистограмма** использует столбцы разной высоты для сравнения распределения числовой переменной между группами категориальной переменной. По умолчанию высота столбца оценивается с помощью «среднего»: параметр estimator изменяет эту функцию агрегирования с помощью встроенных функций Python, таких как estimator=max или len, или функций NumPy наподобие np.max и np.median.

```
# ИСПОЛЬЗОВАНИЕ SEABORN
plt.figure(figsize=(10,6))  # Зададим размер графика
plt.title('Количество супермаркетов по способам оплаты')
plt.xlabel('Способ оплаты')
plt.ylabel('Количество супермаркетов')
sns.histplot(df['Payment'])
```

**Диаграмма рассеяния** показывает взаимосвязь между двумя числовыми признаками с помощью точек, показывающих движение этих переменных вместе.

```
plt.scatter(df['Unit price'], df['Quantity'])
```

```
plt.scatter(df['Unit price'], df['Tax 5%'])
```

**Парная диаграмма** для сравнения распределения пар числовых переменных создаёт сетку точечных диаграмм. Он также содержит гистограмму для каждой функции в диагональных прямоугольниках.

```
sns.pairplot(df[['Unit price','Quantity','gross income','cogs', 'Rating']])
```

**Тепловая карта (heatmap)** — графическое представление значений в сетке цветовой кодировкой. Нанесённые значения — это коэффициенты корреляции пар, отражающие меру линейных отношений, поэтому диаграмма идеальна для пары.

```
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), cmap='coolwarm', linecolor='black', linewidths=1)
```

**Ящик с усами** (boxplot) визуализирует распределение между числовыми и категориальными переменными, отображая информацию о квартилях.
